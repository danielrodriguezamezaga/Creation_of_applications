```python
from jupyter_dash import JupyterDash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import plotly.express as px

df = quandl.get("WIKI/GOOGL", start_date="2015-1-1", end_date="2018-12-31")
# print(df)
figura = px.line(df, title="Cotizacin de Google: ENERO 2015 - DICIEMBRE 2018")
```

```python
app = JupyterDash(__name__)
app.layout = html.Div(children=[
    html.H1(children="APLICACIN CON DASH PARA MERCADOS FINANCIEROS"),
    dcc.Graph(figure=figura)])
if __name__ == "__main__":
    app.run_server(mode="inline")
```
